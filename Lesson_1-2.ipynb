```python
import numpy as np

a = np.array([[1, 6], [2, 8], [3, 11], [3, 10], [1, 7]], int)

print('Массив Numpy:')
print(a)

mean_a = np.mean(a, axis = 0)
print(f'mean_a = {mean_a}')
```

```python
a_centered = a - mean_a
print(a_centered)
```

```python
column_1 = a_centered[:,0]
column_2 = a_centered[:,1]
a_centered_sp = np.dot(column_1, column_2)
covariation = a_centered_sp/(a_centered.shape[0] - 1)
print(f'Ковариация двух признаков: {covariation}')

print(f'Результат через ковариационную матрицу: {np.cov(a.transpose())[0][1]}')
```
