```python
t = np.linspace(0,10,51)
```

```python
plt.plot(t,f, color='g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.xlim(0.5,9.5)
plt.ylim(-2.5,2.5)
plt.show()
```

```python
x = np.linspace(-3,3,51)
```

```python
#y1 = x**2 y2 = 2 * x + 0.5 y3 = -3 * x - 1.5 y4 = sin(x)
y1 = x*x
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x,y1)
ax1.set_title('График_y1')
ax2.set_xlim([-5, 5])
ax2.plot(x,y2)
ax2.set_title('График_y2')
ax3.plot(x,y3)
ax3.set_title('График_y3')
ax4.plot(x,y4)
ax4.set_title('График_y4')
```

```python
url = 'C://Users//boln//creditcard.csv'
df = pd.read_csv(url)
```

```python
class_value_counts = df['Class'].value_counts()
class_value_counts.plot(kind = 'bar')
plt.show()
```

```python
vcv1_class0 = df.loc[df['Class'] == 0,:]['V1'].value_counts()
vcv1_class1 = df.loc[df['Class'] == 1,:]['V1'].value_counts()
```

```python
plt.hist(vcv1_class0, bins = 20, color="lightgrey", ec="white", density =True, alpha = 0.5)
plt.hist(vcv1_class1, bins = 20, color="red", ec="black", density =True, alpha = 0.5)
plt.legend(labels=["Class0", "Class1"])
plt.xlabel("Class")
plt.show()
```
