```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
%matplotlib inline
```

```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.describe()
```

```python
train_data.info()
train_data.head()
```

```python
train_data_norm = train_data.copy()
```

```python
train_data_norm['Ecology_2'] = train_data_norm['Ecology_2'].apply(lambda x: train_data_norm['Ecology_2'].value_counts().to_dict()[x])
```

```python
train_data_norm['Ecology_3'] = train_data_norm['Ecology_3'].apply(lambda x: train_data_norm['Ecology_3'].value_counts().to_dict()[x])
```

```python
train_data_norm['Shops_2'] = train_data_norm['Shops_2'].apply(lambda x: train_data_norm['Shops_2'].value_counts().to_dict()[x])
```

```python
train_data_norm.fillna(-999,inplace=True)
train_data_norm.info()
train_data_norm.head()
```

```python
train_data_norm.info()
train_data_norm.head()
```

```python
X = train_data_norm.iloc[:,0:19]
```

```python
Y = train_data_norm.iloc[:,19:20]
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
regr = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
regr.fit(X_train,Y_train.values[:,0])
```

```python
Y_pred = regr.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
test_data.info()
```

```python
test_data_norm = test_data.copy()
```

```python
test_data_norm['Ecology_2'] = test_data_norm['Ecology_2'].apply(lambda x: test_data_norm['Ecology_2'].value_counts().to_dict()[x])
test_data_norm['Ecology_3'] = test_data_norm['Ecology_3'].apply(lambda x: test_data_norm['Ecology_3'].value_counts().to_dict()[x])
test_data_norm['Shops_2'] = test_data_norm['Shops_2'].apply(lambda x: test_data_norm['Shops_2'].value_counts().to_dict()[x])
test_data_norm.fillna(-999,inplace=True)
test_data_norm.info()
test_data_norm.head()
```

```python
test_pred = regr.predict(test_data_norm)
```

```python
test_pred
```

```python
test_pred_df = pd.DataFrame(test_pred,columns=['Price'])
```
